Exploratory Data Analysis

```python
#plotting the count of loan status of males vs females
sns.countplot(x=df.Gender, hue=df.Loan_Status)
plt.show()
```

Conclusion: Males applied more for loans than females and hence they had a high loan status

```python
#plotting the Loan Amount Vs Graduated and non graduated male female
sns.barplot(x="Gender", y="LoanAmount", hue="Education", data=df, width=0.7)
plt.show()
```

A bar graph which shows Graduated Males took more loans than Graduated Females

```python
#plotting the count of loan status of people living in different areas
sns.countplot(x=df.Property_Area, hue=df.Loan_Status)
plt.show()
```

Conclusion: People in Semi-urban Areas took more loans than in Urban and Rural Areas

```python
#plotting a scatter plot of loan amount vs loan amount term among male and female
sns.scatterplot(x="LoanAmount", y="Loan_Amount_Term", hue="Gender", data=df, style="Gender")
plt.show()
```
